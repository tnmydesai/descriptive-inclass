```python
import numpy as np
Time_spent=[5.6,  6.2,  6.8, 7.6, 7.9 ,8.2 ,8.9 ,9.7 ,9.8, 9.8, 10, 10.1, 10.5, 13.5, 13.7]
print("The standard deviation of each persons's time spent on busy day",np.std(Time_spent),"hours")
```

```python
data=[10,15,20,25,30,35,40,45,50,55,60]

q1=np.quantile(data,q=0.25)
q2=np.quantile(data,q=0.50)
q3=np.quantile(data,q=0.75)

IQR= (q3-q1)
print("The inter quartile range is",IQR)
```

```python
Q3=15000
Q1=5000

Quartile_Deviation=(Q3-Q1)/2

Coeff_of_Quartile_Deviation=(Q3-Q1)/(Q3+Q1)

print("The absolute measure of dispersion is",Quartile_Deviation)
print("The relative measure of dispersion is",Coeff_of_Quartile_Deviation)
```
