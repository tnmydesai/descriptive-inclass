```python
# Creating a dataframe with the values
import pandas as pd
import numpy as np

df=pd.DataFrame(data={'Emp_No':[1,3,2,4,5,6,7,8,9,10],'Monthly_Salary':[90,18,80,18,17,16,16,16,15,14]})
df
```

```python
print("The Average Monthly salary of the Employees",df['Monthly_Salary'].mean())
```

```python
print("The Mean of the Monthly Salary",df['Monthly_Salary'].mean())
print("The Median of the Monthly Salary",df['Monthly_Salary'].median())
```

```python
print("The most frequent salary of the employess is",df['Monthly_Salary'].mode()[0])
```

```python
df.aggregate({'Monthly_Salary':['var','std']})
```

```python
# First we are going to find the range
range_max=df['Monthly_Salary'].max()
range_min=df['Monthly_Salary'].min()
range_MS=(range_max-range_min)
print("The range of the Monthly salary",range_MS)
```

```python
# Second we are going to compute Variance
Var_MS=df['Monthly_Salary'].var()
print("The Variance of the Monthly salary",Var_MS)
```

```python
# we are going to compute the Standard deviation
std_MS=df['Monthly_Salary'].std()
print("The standard deviation of the Monthly salary",std_MS)
```

```python
# Last the interquartile range
q1=df['Monthly_Salary'].quantile(0.25)
q3=df['Monthly_Salary'].quantile(0.75)

# The inter quartile range of the variable

iqr=q3-q1

print("The Interquartile range of the variable is",iqr)
```
